```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/cleaned_adult.csv')
df.head()
```

```python
# boxplot is based on Inter Quantile Range
Q1 = df['age'].quantile(.25)
Q3 = df['age'].quantile(.75)

IQR = Q3 - Q1

LL = Q1-1.5*IQR
UL = Q3+1.5*IQR


median = df['age'].median()

df['age'] = np.where( df['age'] > UL , median, df['age'])
df['age'] = np.where( df['age'] < LL , median, df['age'])
```

```python
df.boxplot('age')
```

```python
# analysis

# filter , sorting , group by , head , pivottable



# filter
df[  df['age'] >50  ]  # single filter on dataset

df [ (df['age'] > 50)    &   (df['race'] == 'White') ]  # multi column filters ---- & and , | or


# sort

df[  (df['age']>50)   &    (df['race']=='White')].sort_values(by = 'age')  # sorting in ascending order

df[  (df['age']>50)   &    (df['race']=='White')].sort_values(by = 'age' , ascending = False)  # descending order

# practice question for filtering and sorting

# Filter all records where the person works more than 40 hours per week. Sort the result by hours-per-week in descending order.
# Filter people whose education is "Bachelors" and whose income is ">50K". Sort them by age from youngest to oldest.
# Filter records where occupation is "Exec-managerial" or "Prof-specialty". Sort the result first by occupation, then by age in descending order.
# Filter people from the United-States whose capital-gain is greater than 0. Sort them by capital-gain from highest to lowest.
# Filter female records where marital-status is "Never-married". Sort the result by education-num in descending order, then by age in ascending order.
```

```python
import pandas as pd

df = pd.read_csv('/content/Bank Customer Churn Prediction.csv')
df.head()
```

```python
df.drop('customer_id' , axis = 1, inplace = True)
```

```python
# feature transformation
# encoding
# scaling

from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()


df.gender = lb.fit_transform(df.gender)
df.country = lb.fit_transform(df.country)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.gender.unique()
```

```python
x = df.drop( 'churn' , axis =1)

cls = df['churn']

```

```python
from sklearn.model_selection import train_test_split

x_train, x_test , y_train , y_test = train_test_split(x, cls , test_size= .2)
```

```python
# regression model

from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression , LogisticRegression
from sklearn.metrics import r2_score , accuracy_score

regresssion = LogisticRegression()

regresssion.fit(x_train , y_train)

pred = regresssion.predict(x_test)

accuracy_score(y_test , pred)
```

```python
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.metrics import r2_score, accuracy_score

regression = LogisticRegression()
regression.fit(x_train,y_train)
pred = regression.predict(x_test)
accuracy_score(y_test,pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

dt = DecisionTreeClassifier(criterion = 'entropy', max_depth = 8)
dt.fit(x_train,y_train)
pred = dt.predict(x_test)
accuracy_score(y_test,pred)
```

```python
from sklearn.tree import plot_tree
plot_tree(dt)
```
